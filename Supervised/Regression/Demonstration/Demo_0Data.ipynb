## Task 3.1: Differencing
-   One-lag differencing: Step h = 1 (perious = 1)
-   Two-lag differening: Step h = 2 (perious = 2)


```python
# Cov (X_t,X_{t+h})??
## 1-lag differencing (perious = 1 rows)
one_lag_diff = df_BMW["Close"].diff(periods=1)
fig, ax = plt.subplots(1,1,figsize = (15,5))
ax.plot(df_BMW["Date"][1:],one_lag_diff[1:]) # We only have diff start from 1
fig.suptitle("Graph of one_lag_diff")
ax.set_xlabel("Date_time")
ax.set_ylabel("Differencing")
ax.set_xticks(ax.get_xticks()[::60]) #Without this it is too dense, with this we only display every after 60 xticks
ax.set_xticklabels(ax.get_xticklabels(),rotation = 40, ha = 'center')# Shearing the labels to 40 degree and write it in center
plt.show()

```

```python
two_lag_diff = df_BMW["Close"].diff(periods=2)
fig, ax = plt.subplots(1,1,figsize = (15,5))
ax.plot(df_BMW["Date"][2:],two_lag_diff[2:]) # We only have two_lag_diff from index 2 
fig.suptitle("Graph of two_lag_diff")
ax.set_xlabel("Date_time")
ax.set_ylabel("Differencing")
ax.set_xticks(ax.get_xticks()[::60]) #Without this it is too dense, with this we only display every after 60 xticks
ax.set_xticklabels(ax.get_xticklabels(),rotation = 40, ha = 'center')# Shearing the labels to 40 degree and write it in center
plt.show()
```

### Task 3.1.1: How to use ADF Test?
-   Test for d = 0,
-   Test for d = 1,

```python
from statsmodels.tsa.stattools import adfuller
# Show ther perform augmented Dickey-Fuller test:
adfuller(df_BMW["Close"])
```

Most important values in the output:
-   Test satistic: -1.849349
-   P-Value: 0.35625

-> p-value is not less than 0.05: So with d = 0 the time series is non-stationary.

In crease d, compute d-lag differencing, apply the adf test

```python
one_lag_diff = df_BMW['Close'].diff(periods = 1).dropna()
adfuller(one_lag_diff)
```

p-value = 0.0 -> one-lag differencing is stationary.

Choose d = 1
Plot ACF and PACF of data to choose p,q:

```python
# Plot auto correlation depends on lag
auto_corr_Close = pd.plotting.autocorrelation_plot(df_BMW["Close"])
auto_corr_Close.plot()
plt.show()
```

```python
import numpy as np
lag_1_50_autocorrs = [df_BMW["Close"].autocorr(lag=i) for i in range(1,51)]
x_axis = np.arange(1,51)
fig, ax = plt.subplots(1,1,figsize = (15,5))
ax.stem(x_axis,lag_1_50_autocorrs) # We only have two_lag_diff from index 2 
fig.suptitle("Autocorrelation function")
ax.set_xlabel("Lags")
ax.set_ylabel("Autocorrelation")
ax.set_xticks(np.arange(1,51))
plt.show()
```
