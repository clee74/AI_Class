```python
# Visualization
from matplotlib import pyplot as plt
import matplotlib
feature_ram = df_mobile["ram"]
feature_intM = df_mobile["int_memory"]
lablel = raw_y
# Make a figure
plot_1 = plt.figure()
# Update figure
colors = ["red","green","blue","orange","yellow"]
ax = plot_1.add_subplot(1,1,1)
scatter = ax.scatter(feature_intM,feature_ram,alpha=0.8,c=lablel,s = 10,cmap=matplotlib.colors.ListedColormap(colors))
legend1 = ax.legend(*scatter.legend_elements(),loc='upper left', title = "Classes")
ax.add_artist(legend1)
ax.set_ylabel("Ram")
ax.set_xlabel("Int_memmory")
plt.title
```
