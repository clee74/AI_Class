# Demo 2: Dimensional Reduction with Pcincipal Component Analysis (PCA).

## Task 1: Read Data

```python
from sklearn import datasets
```

```python
# Load data
digits = datasets.load_digits()
print(digits.data.shape)
```

```python
import numpy as np
n_samples = len(digits.images)
data = digits.images.reshape((n_samples,-1))
# Normalization 
# Why devided by 16 not 255? max min
X = data/16
y_labels = digits.target
y_labels = np.array(y_labels)
```

## Task 2: PCA to reduce the number of dimension 

```python
from sklearn.decomposition import PCA
import pandas as pd

import time
from matplotlib import pyplot as plt
```

```python
# number of dimension in output = 2
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)
```

```python
# Visualize results:

fig,ax = plt.subplots(1,1,figsize = (10,6))
scatter = ax.scatter(pca_result[:,0],pca_result[:,1],c = y_labels)

ax.legend(*scatter.legend_elements(),loc = 'upper left',title="Clusters")
```
