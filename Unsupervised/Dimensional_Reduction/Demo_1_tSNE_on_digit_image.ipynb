# Demo 1: Dimensional Reduction with T-Distributed Stochastic Neighbor Embedding (tSNE)

## Task 1: Read Data

```python
from sklearn import datasets
```

```python
# Load data
digits = datasets.load_digits()
print(digits.data.shape)
```

```python
import numpy as np
n_samples = len(digits.images)
data = digits.images.reshape((n_samples,-1))
# Normalization 
# Why devided by 16 not 255? max min
X = data/16
y_labels = digits.target
y_labels = np.array(y_labels)
```

## Task 2: TSNE to Reduce the number of dimension

```python
from sklearn.manifold import TSNE
import pandas as pd

import time
from matplotlib import pyplot as plt
```

```python
time_start = time.time()
tsne = TSNE(n_components=2, verbose = 1, perplexity=40,n_iter = 300)
tsne_results = tsne.fit_transform(X)
```

```python
# Visualize results:

fig,ax = plt.subplots(1,1,figsize = (10,6))
scatter = ax.scatter(tsne_results[:,0],tsne_results[:,1],c = y_labels)

ax.legend(*scatter.legend_elements(),loc = 'upper left',title="Clusters")
```
