# Task 1: psudo dataset

```python
from sklearn.datasets import make_classification
data,_ = make_classification(
    n_samples=1000,
    n_features=2,
    n_classes=3,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4,
    class_sep=2
)
```

# Task 2: Hierachical Clusstering: Agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
model_agg = AgglomerativeClustering(n_clusters=None,distance_threshold=10)
# Fitting data to kmeans
model_agg.fit(data)
# Predict clusters class for each data point
pred_agg = model_agg.fit_predict(data)
```

```python
from numpy import unique,where
from matplotlib import pyplot as plt
kmean_clusters = unique(pred_agg)

fig,ax = plt.subplots(1,1,figsize = (5,5))
scatter = ax.scatter(data[:,0],data[:,1],c = pred_agg)

ax.legend(*scatter.legend_elements(),loc = 'upper left',title="Clusters")
```

# Dendrogram charts
The way to find the optimal number of clusters in hierarchical clustering is to use a dendrogram chart.

```python
from scipy.cluster.hierarchy import dendrogram,linkage

plt.figure(figsize=(16,5))
plt.title("Dendrogram")
plt.ylabel("Euclidean distances")
plt.xlabel("Data points")
link_matrix = linkage(data,method='ward')
dendrogram = dendrogram(link_matrix)
plt.show()
```
