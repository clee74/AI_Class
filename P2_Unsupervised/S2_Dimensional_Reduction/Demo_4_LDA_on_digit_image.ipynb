# Demo 4: Dimensional Reduction with Latent Dirichlet Allocation

## Task 1: Read Data

```python
from sklearn import datasets
```

```python
# Load data
digits = datasets.load_digits()
print(digits.data.shape)
```

```python
import numpy as np
n_samples = len(digits.images)
data = digits.images.reshape((n_samples,-1))
# Normalization 
# Why devided by 16 not 255? max min
X = data/16
y_labels = digits.target
y_labels = np.array(y_labels)
```

## Task 2: LDA to reduce the number of dimension 

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
import pandas as pd
import time
from matplotlib import pyplot as plt
```

```python
lda = LDA(n_components=3)
lda_result = lda.fit_transform(X)
```

```python
# Visualize results:

fig,ax = plt.subplots(1,1,figsize = (10,6))
scatter = ax.scatter(lda_result[:,0],lda_result[:,1],c = y_labels)

ax.legend(*scatter.legend_elements(),loc = 'upper left',title="Clusters")
```
