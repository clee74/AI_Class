# Demo 3: Dimensional Reduction with Latent Semantic Analysis (LSA)

## Task 1: Read Data

```python
from sklearn import datasets
```

```python
# Load data
digits = datasets.load_digits()
print(digits.data.shape)
```

```python
import numpy as np
n_samples = len(digits.images)
data = digits.images.reshape((n_samples,-1))
# Normalization 
# Why devided by 16 not 255? max min
X = data/16
y_labels = digits.target
y_labels = np.array(y_labels)
```

## Task 2: LSA to reduce the number of dimension

```python
import pandas as pd

import time
from matplotlib import pyplot as plt
from sklearn.decomposition import TruncatedSVD
```

```python

lsa = TruncatedSVD(2,algorithm='arpack')
lsa_result = lsa.fit_transform(X)
```

```python
# Visualize results:

fig,ax = plt.subplots(1,1,figsize = (10,6))
scatter = ax.scatter(lsa_result[:,0],lsa_result[:,1],c = y_labels)

ax.legend(*scatter.legend_elements(),loc = 'upper left',title="Clusters")
```
